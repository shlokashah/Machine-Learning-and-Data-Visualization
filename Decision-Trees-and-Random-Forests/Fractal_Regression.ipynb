```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('onlinenews_train.csv')
data.drop('RowID',axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['n_tokens_content'] = scaler.fit_transform(data[['n_tokens_content']])
data['n_tokens_title'] = scaler.fit_transform(data[['n_tokens_title']])
data.head()
```

```python
data.columns
```

```python
data.corr()
```

```python
data.drop('url',axis=1,inplace=True)
```

```python
features=['n_tokens_title','n_tokens_content','n_unique_tokens','n_non_stop_words','n_non_stop_unique_tokens','num_hrefs',
         'num_self_hrefs','num_imgs','num_videos','num_keywords','data_channel_is_lifestyle','data_channel_is_socmed','kw_max_min',
         'kw_avg_min','kw_min_max','kw_max_max','kw_avg_max','kw_avg_max','kw_min_avg', 'kw_max_avg',
       'kw_avg_avg', 'self_reference_min_shares', 'self_reference_max_shares',
       'self_reference_avg_sharess', 'weekday_is_monday','weekday_is_saturday', 'weekday_is_sunday', 'is_weekend', 'LDA_00',
       'LDA_01', 'LDA_02', 'LDA_03','global_subjectivity','global_sentiment_polarity','global_rate_positive_words',
       'global_rate_negative_words','rate_positive_words',
       'rate_negative_words','avg_positive_polarity','max_positive_polarity','title_subjectivity',
       'title_sentiment_polarity', 'abs_title_subjectivity',
       'abs_title_sentiment_polarity']
X = data[features]
```

```python
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer.fit(X)
X = imputer.transform(X)
# data['n_tokens_content'] = normalize(data[['n_tokens_content']])
```

```python
y = data['shares'].fillna(data['shares'].mean())
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(X_train,y_train)
```

```python
pred = linear.predict(X_test)
```

```python
from sklearn import metrics
```

```python
plt.scatter(y_test,pred)
```

```python
sns.distplot((y_test-pred),bins=50);
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(pred,y_test)))
```

```python
r_score = linear.score(X_test,y_test)
r_score
```
