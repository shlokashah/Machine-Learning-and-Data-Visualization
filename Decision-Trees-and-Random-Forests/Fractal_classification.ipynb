```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('letterdata_train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['xbox'].fillna(data['xbox'].mean(),inplace=True)
data['ybox'].fillna(data['ybox'].mean(),inplace=True)
data['width'].fillna(data['width'].mean(),inplace=True)
data['height'].fillna(data['height'].mean(),inplace=True)
data['onpix'].fillna(data['onpix'].mean(),inplace=True)
data['xbar'].fillna(data['xbar'].mean(),inplace=True)
data['ybar'].fillna(data['ybar'].mean(),inplace=True)
data['x2bar'].fillna(data['x2bar'].mean(),inplace=True)
data['y2bar'].fillna(data['y2bar'].mean(),inplace=True)
data['xybar'].fillna(data['xybar'].mean(),inplace=True)
data['x2ybar'].fillna(data['x2ybar'].mean(),inplace=True)
data['xy2bar'].fillna(data['xy2bar'].mean(),inplace=True)
data['xedge'].fillna(data['xedge'].mean(),inplace=True)
data['xedgey'].fillna(data['xedgey'].mean(),inplace=True)
data['yedge'].fillna(data['yedge'].mean(),inplace=True)
data['yedgex'].fillna(data['yedgex'].mean(),inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data.dropna(inplace=True)
rfc = RandomForestClassifier(n_estimators=200)
```

```python
X = data.drop('letter',axis=1)
y = data['letter']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
rfc.fit(X_train,y_train)
```

```python
data_test = pd.read_csv("letterdata_test.csv")
data_test
```

```python
data_test.dropna(inplace=True)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
rfc_pred2 = rfc.predict(data_test)
```
